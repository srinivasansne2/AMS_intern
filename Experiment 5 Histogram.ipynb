# HISTOGRAM OF COLOR IMAGE

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img=cv.imread(r"C:\Users\ASUS\Desktop\Fort.jpg")
src1 = cv.cvtColor(img, cv.COLOR_BGR2RGB)


color = ('r','g','b')

for channel,col in enumerate(color):
    histr = cv.calcHist([src1],[channel],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
    plt.title('Histogram of Color Image')

plt.show()
plt.imshow(src1)
plt.show
```
