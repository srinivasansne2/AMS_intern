## Conversion of Color Image to Grayscale & Binary Image

```python
image = cv2.imread(r'C:\Users\ASUS\Desktop\ROSE.jpeg')
RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
width = 480
height = 480
dim = (width, height)  

resized = cv2.resize(RGB, dim, interpolation = cv2.INTER_AREA)
```

```python
grayImage = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
  
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(resized)
```

```python
plt.imshow(blackAndWhiteImage)
```

## Identifying Color

```python
img2 = cv2.imread(r'C:\Users\ASUS\Desktop\ROSE.jpeg')
```

```python
# Convert to hsv colorpace 
hsv = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)

# lower bound and upper bound for Yellow color
lower_bound = np.array([0,50,50])
upper_bound = np.array([10,255,255])

# find the colors within the boundaries
mask = cv2.inRange(hsv, lower_bound, upper_bound)
```

```python
# Display Image and Mask
cv2.imshow("Original Image", img2)
cv2.imshow("Mask", mask)
```
