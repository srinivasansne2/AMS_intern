## IMAGE RESIZING

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the image
image = cv2.imread(r"C:\Users\ASUS\Desktop\ROSE.jpeg")
```

```python
#convert the image to RGB (images are read in BGR in OpenCV)
RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
print('Original Dimensions : ',RGB.shape)
```

```python
half = cv2.resize(RGB, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(RGB, (1050, 1610))
stretch_near = cv2.resize(RGB, (780, 540),interpolation = cv2.INTER_NEAREST)
```

```python
Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[RGB, half, bigger, stretch_near]
count = 4
```

```python
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
    
plt.show()
```

```python
print('Original Dimensions : ',RGB.shape)
print('Resized Dimensions : ',half.shape)
print('Resized Dimensions : ',bigger.shape)
print('Resized Dimensions : ',stretch_near.shape)
```

# Resize only width

```python
import cv2
```

```python
# Loading the image
image = cv2.imread(r"C:\Users\ASUS\Desktop\ROSE.jpeg")
```

```python
#convert the image to RGB (images are read in BGR in OpenCV)
RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
print('Original Dimensions : ',RGB.shape)
```

```python
width = 440
height = RGB.shape[0]  # keep original height
dim = (width, height)
```

```python
# resize image
resized = cv2.resize(RGB, dim, interpolation = cv2.INTER_AREA)
```

```python
print('Resized Dimensions : ',resized.shape)
```

```python
cv2.imshow("Resized image", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
